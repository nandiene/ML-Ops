```python
#import libraries
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read the image
image = cv2.imread('images/tomato farm.jpg')
image.shape
#write the image
image_in = cv2.imwrite('images/tomato farm_new.jpg', image)
#visualize the image
cv2.imshow('Tomato Farm', image)
cv2.waitKey(20)
```

```python
#visualizing using matplotlib
plt.imshow(image)
```

```python
#read and visualizing the video

videol = cv2.VideoCapture('images/homeless.mp4')#reading a video
ret = True
while ret:
    ret, frame = videol.read()
    if ret:
        cv2.imshow('frame', frame)
        cv2.waitKey(25)
videol.release()
cv2.destroyAllWindows()
```

```python
#read and visualizing a webcam
webcam = cv2.VideoCapture(0)#read webcam
#visualize webcam
ret = True
while ret:
    ret, frame = webcam.read()
    cv2.imshow('frame', frame)
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break
webcam.release()
cv2.destroyAllWindows()
```

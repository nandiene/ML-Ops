```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import dataset
salary_data= pd.read_csv('salary_data.csv')
```

```python
#extract dependent and independent variables
X = salary_data.iloc[:, :-1].values
y = salary_data.iloc[:, 1].values
```

```python
#visualize the data
sns.displot(salary_data['Years of Experience'],kde= False, bins=5 )
```

```python
sns.heatmap(salary_data.corr())
```

```python
sns.countplot(y= 'Years of Experience',data= salary_data)
```

```python
sns.barplot(data= salary_data, x='Years of Experience', y= 'Salary')
```

```python
# Split the data into training set and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)
```

```python
#Fitting the Linear regression model into training set
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
#predicting the test results
y_pred =lr.predict(X_test)
y_pred
```

```python
#plot the test set
plt.scatter(X_test,y_test, color= 'green')
```

```python
# Plot the training set results
plt.scatter(X_train, y_train, color= 'blue')
plt.plot(X_train,lr.predict(X_train), color='red')
plt.title('Salary - Year of Experience(Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
